```python
from PIL import Image
import os
```

```python
def convert_images_to_jpg(input_folder, output_folder):
    # Tạo folder đầu ra nếu chưa tồn tại
    os.makedirs(output_folder, exist_ok=True)

    for filename in os.listdir(input_folder):
        input_path = os.path.join(input_folder, filename)

        # Bỏ qua nếu không phải file
        if not os.path.isfile(input_path):
            continue
        
        # Lấy tên file không có phần mở rộng
        name, ext = os.path.splitext(filename)

        # Load ảnh
        try:
            with Image.open(input_path) as img:
                # Đảm bảo ảnh được mở dưới dạng RGB
                img = img.convert("RGB")

                # Đường dẫn lưu file JPG
                output_path = os.path.join(output_folder, f"{name}.jpg")

                # Lưu ảnh dưới định dạng JPG
                img.save(output_path, format="JPEG", quality=100)
                # print(f"Converted: {filename} -> {output_path}")

        except Exception as e:
            print(f"Error processing {filename}: {e}")
```

```python
# Đường dẫn folder input và output
input_folder = "datasets/pa3.1_malaria_7_classes/train/images"  # Thay bằng folder ảnh gốc
output_folder = "datasets/pa3.1_malaria_7_classes/train2/images"  # Thay bằng folder lưu ảnh JPG

convert_images_to_jpg(input_folder, output_folder)
```
